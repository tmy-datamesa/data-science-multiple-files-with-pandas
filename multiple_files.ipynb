```python
import numpy as np
import pandas as pd
import matplotlib
```

## Merge AlÄ±ÅŸtÄ±rmasÄ±

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### Ä°Ã§ BirleÅŸtirme (inner merge)

```python
inner_merged_df = a_df.merge(b_df, on='Country', how='inner')
inner_merged_df
```

her iki DataFrame'de de ortak olan anahtar deÄŸerlere sahip satÄ±rlarÄ± birleÅŸtirir. Ortak olmayan satÄ±rlar sonuÃ§ DataFrame'inde yer almaz. Bu iÅŸlemde Finland ve Canada satÄ±rlarÄ±, diÄŸer DataFrame'de karÅŸÄ±lÄ±klarÄ± olmadÄ±ÄŸÄ± iÃ§in sonuÃ§ta yer almayacaktÄ±r.

```python
from nbresult import ChallengeResult

result = ChallengeResult('inner_merge',
    inner_merged_shape=inner_merged_df.shape,
    inner_merged_nulls=sum(inner_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

## Sol BirleÅŸtirme (Left Merge)

```python
inner_merged_df = a_df.merge(b_df, on='Country', how='left')
inner_merged_df
```

```python
left_merged_df = pd.merge(a_df, b_df, on='Country', how='left')
left_merged_df
```

Sol birleÅŸtirme, sol DataFrame'deki tÃ¼m satÄ±rlarÄ± korur ve saÄŸ DataFrame'den eÅŸleÅŸen satÄ±rlarÄ± ekler. SaÄŸ DataFrame'de eÅŸleÅŸme bulunamayan yerlere NaN (Not a Number) deÄŸerleri atanÄ±r. Bu durumda Finland satÄ±rÄ± korunacak, ancak HDI deÄŸeri NaN olacaktÄ±r. Canada ise sonuÃ§ta yer almayacaktÄ±r.

```python
from nbresult import ChallengeResult

result = ChallengeResult('left_merge',
    left_merged_shape=left_merged_df.shape,
    left_merged_nulls=sum(left_merged_df.isnull().sum())
)
result.write()

print(result.check())
```
