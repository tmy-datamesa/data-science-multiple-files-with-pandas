```python
import numpy as np
import pandas as pd
import matplotlib
```

## Merge AlÄ±ÅŸtÄ±rmasÄ±

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

### Ä°Ã§ BirleÅŸtirme (inner merge)

```python
inner_merged_df = a_df.merge(b_df, on='Country', how='inner')
inner_merged_df
```

her iki DataFrame'de de ortak olan anahtar deÄŸerlere sahip satÄ±rlarÄ± birleÅŸtirir. Ortak olmayan satÄ±rlar sonuÃ§ DataFrame'inde yer almaz. Bu iÅŸlemde Finland ve Canada satÄ±rlarÄ±, diÄŸer DataFrame'de karÅŸÄ±lÄ±klarÄ± olmadÄ±ÄŸÄ± iÃ§in sonuÃ§ta yer almayacaktÄ±r.

```python
from nbresult import ChallengeResult

result = ChallengeResult('inner_merge',
    inner_merged_shape=inner_merged_df.shape,
    inner_merged_nulls=sum(inner_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

## Sol BirleÅŸtirme (Left Merge)

```python
inner_merged_df = a_df.merge(b_df, on='Country', how='left')
inner_merged_df
```

```python
left_merged_df = pd.merge(a_df, b_df, on='Country', how='left')
left_merged_df
```

Sol birleÅŸtirme, sol DataFrame'deki tÃ¼m satÄ±rlarÄ± korur ve saÄŸ DataFrame'den eÅŸleÅŸen satÄ±rlarÄ± ekler. SaÄŸ DataFrame'de eÅŸleÅŸme bulunamayan yerlere NaN (Not a Number) deÄŸerleri atanÄ±r. Bu durumda Finland satÄ±rÄ± korunacak, ancak HDI deÄŸeri NaN olacaktÄ±r. Canada ise sonuÃ§ta yer almayacaktÄ±r.

```python
from nbresult import ChallengeResult

result = ChallengeResult('left_merge',
    left_merged_shape=left_merged_df.shape,
    left_merged_nulls=sum(left_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

## SaÄŸ BirleÅŸtirme (Right Merge)

```python
right_merged_df = pd.merge(a_df, b_df, on='Country', how='right')
right_merged_df
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('right_merge',
    right_merged_shape=right_merged_df.shape,
    right_merged_nulls=sum(right_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

## DÄ±ÅŸ BirleÅŸtirme (Outer Merge)

```python
outer_merged_df = pd.merge(a_df, b_df, on='Country', how='outer')
outer_merged_df.sort_values(by='Population (M)', ascending=False)
```

DÄ±ÅŸ birleÅŸtirme, her iki DataFrame'deki tÃ¼m satÄ±rlarÄ± korur. EÅŸleÅŸme bulunamayan yerlere NaN deÄŸerleri atanÄ±r. Bu durumda hem Finland hem de Canada satÄ±rlarÄ± sonuÃ§ta yer alacak ve eksik deÄŸerler NaN ile doldurulacaktÄ±r.

```python
from nbresult import ChallengeResult

result = ChallengeResult('outer_merge',
    outer_merged_shape=outer_merged_df.shape,
    outer_merged_nulls=sum(outer_merged_df.isnull().sum())
)
result.write()

print(result.check())
```

# Join AlÄ±ÅŸtÄ±rmasÄ±

join() fonksiyonu, merge()'e benzer ancak varsayÄ±lan olarak DataFrame'lerin indeksleri Ã¼zerinden birleÅŸtirme yapar. Genellikle bir DataFrame'in indeksini baÅŸka bir DataFrame'in sÃ¼tunuyla birleÅŸtirmek veya iki DataFrame'i indeksleri Ã¼zerinden birleÅŸtirmek iÃ§in kullanÄ±lÄ±r.

Merge: DeÄŸer bazlÄ±dÄ±r ($f(value_A, value_B)$).Join: Adres (Ä°ndeks) bazlÄ±dÄ±r ($f(index_A, index_B)$).

```python
aa_df = a_df.set_index("Country")
aa_df
```

```python
bb_df = b_df.set_index("Country")
bb_df
```

```python
aa_df.join(bb_df)
```

VarsayÄ±lan olarak .join() sol (left) birleÅŸtirme yapar. DiÄŸer birleÅŸtirme tÃ¼rlerini de how parametresi ile deneyebilirsiniz:

```python
aa_df.join(bb_df, how='inner')
```

```python
aa_df.join(bb_df, how='right')
```

```python
aa_df.join(bb_df, how='outer')
```

Ne zaman merge() ne zaman join() kullanmalÄ±yÄ±z?

Belirli bir sÃ¼tuna gÃ¶re birleÅŸtirmek istediÄŸinizde merge(); indeks Ã¼zerinden birleÅŸtirmek istediÄŸinizde join() kullanabilirsiniz.



## Concat AlÄ±ÅŸtÄ±rmasÄ±

```python
concat_df = pd.concat([a_df, b_df], axis="index", sort=False)
concat_df
```
